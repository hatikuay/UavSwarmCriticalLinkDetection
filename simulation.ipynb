```python
import pandas as pd

# Load the CSV results
df = pd.read_csv('parameter_sweep_results.csv')

# Compute reward as mean_lambda2 - 0.1 * partition_rate
df['reward'] = df['mean_lambda2'] - 0.1 * df['partition_rate']

# Find the best parameter set
best = df.loc[df['reward'].idxmax()]

best_params = {
    'alpha': best['alpha'],
    'beta': best['beta'],
    'gamma': best['gamma'],
    'mean_lambda2': best['mean_lambda2'],
    'partition_rate': best['partition_rate'],
    'reward': best['reward']
}

best_params
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the results
df = pd.read_csv('parameter_sweep_results.csv')
df['reward'] = df['mean_lambda2'] - 0.1 * df['partition_rate']

# Plot Mean Lambda2 vs Alpha
plt.figure()
plt.scatter(df['alpha'], df['mean_lambda2'])
plt.xlabel('Alpha')
plt.ylabel('Mean Lambda2')
plt.title('Mean Algebraic Connectivity vs Alpha')
plt.show()

# Plot Partition Rate vs Beta
plt.figure()
plt.scatter(df['beta'], df['partition_rate'])
plt.xlabel('Beta')
plt.ylabel('Partition Rate')
plt.title('Partition Rate vs Beta')
plt.show()

# Plot Reward vs Gamma
plt.figure()
plt.scatter(df['gamma'], df['reward'])
plt.xlabel('Gamma')
plt.ylabel('Reward (Mean Lambda2 - 0.1*Partition Rate)')
plt.title('Reward vs Gamma')
plt.show()
```
